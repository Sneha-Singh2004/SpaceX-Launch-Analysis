```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("../data/cleaned_spacex_data.csv")
df.head()
```

```python
conn = sqlite3.connect("spacex.db")
df.to_sql("spacex", conn, if_exists="replace", index=False)
```

```python
query = """
SELECT strftime('%Y', date_utc) AS year, COUNT(*) AS total_launches
FROM spacex
GROUP BY year
ORDER BY year;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT success, COUNT(*) as count
FROM spacex
GROUP BY success;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT strftime('%Y', date_utc) AS year,
       AVG(success) AS success_rate
FROM spacex
GROUP BY year
ORDER BY year;
"""

pd.read_sql(query, conn)
```
