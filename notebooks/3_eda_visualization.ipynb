```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
```

```python
df = pd.read_csv("../data/cleaned_spacex_data.csv")
df.head()
```

```python
sns.countplot(x='success', data=df)
plt.title("Launch Success Distribution")
plt.show()
```

```python
df['date_utc'] = pd.to_datetime(df['date_utc'])
df['year'] = df['date_utc'].dt.year
```

```python
df['year'].value_counts().sort_index().plot(kind='bar')
plt.title("Launches Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Launches")
plt.show()
```

```python
df.groupby('year')['success'].mean().plot(marker='o')
plt.title("Success Rate Over Years")
plt.ylabel("Success Rate")
plt.show()
```
